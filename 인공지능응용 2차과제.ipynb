이미지를 위한 패키지 다운로드

```python
import cv2
from google.colab.patches import cv2_imshow
```

이미지를 첨부한 뒤 글자를 제대로 인식하는지 테스트

```python
import easyocr
file="name.png"
reader = easyocr.Reader(['ko'], gpu=False)
img = cv2.imread(file)
cv2_imshow(img)
text = reader.readtext(img, detail=0)
print(text)
```
